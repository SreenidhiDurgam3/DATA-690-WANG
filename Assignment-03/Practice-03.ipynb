```python
from random import randint #using randint from random library
for j in range(10): 
   matr_list=[]  #an empty list to store values
   for i in range(10):
     value=randint(0,9)
     matr_list.append(value) #appending values in list
 
   res = [str(a) for a in matr_list]  #storing values of list in a matrix
   print(" " . join(res)) 
```

```python
from random import randint
for j in range(10):
   matr_list=[]   
   for i in range(10):
     value=randint(0,9)
     if(value%2!=0): #if condition for value is an odd number
         value='@'  #replace value with '@'
     matr_list.append(value)
 
   sarr = [str(a) for a in matr_list]
   print(" " . join(sarr))
```

```python
#Calculate and Print the total of each row like the following (use "*" to separate the total from the numbers):
from random import randint
for j in range(10):
   matr_list=[]
   row_sum=0   
   for i in range(10):
     value=randint(0,9)
     matr_list.append(value)
     row_sum=row_sum+value  #row sum as sum of all values in each row
   matr_list.append("*")
   matr_list.append(row_sum)
 
   res = [str(a) for a in matr_list]
   print(" " . join(res))
```
